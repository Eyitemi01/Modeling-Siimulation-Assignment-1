```python
def infinite_queue(arrival_rate, service_rate, n):
    rho = arrival_rate / service_rate
    LenSystem = rho / (1 - rho)
    LenQueue = LenSystem - rho
    WaitTimeSystem = LenSystem / arrival_rate
    WaitTimeQueue = LenQueue / arrival_rate


    print(f"The expected length of the system is {round(LenSystem, 3)} people")
    print(f"The expected length of the queue is {round(LenQueue, 3)} people")
    print(f"The expected waiting time of the system is {WaitTimeSystem} hours")
    print(f"The expected waiting time of the queue is {WaitTimeQueue} hours")
    
    option = int(input(f"Are you looking for the probability of 1.exactly 2.At least 3.At most {n} people? (select corresponding numbers)"))
    if option == 1:
        print(f"The probability of {n} people in the system is {(prob_of_exact(n, rho))*100}%")
    elif option == 2:
        print(f"The probability of at least {n} people in the system is {(prob_of_atleast(n, rho))*100}%")
    elif option == 3:
        print(f"The probability of at most {n} people in the system is {(prob_of_atmost(n, rho))*100}%")


def finite_queue(arrival_rate, service_rate, n, N):
    rho = arrival_rate / service_rate
    LenSystem = (rho * ((1 + (N * (rho ** (N + 1)))) - ((N + 1) * (rho ** N))) / ((1 - rho) * (1 - (rho ** (N + 1)))))
    arr_eff = arrival_rate * (1 - prob_of_exact(n, rho, N))
    LenQueue = LenSystem - (arr_eff / service_rate)
    WaitTimeSystem = LenSystem / arr_eff
    WaitTimeQueue = LenQueue / arr_eff

    print(f"The expected length of the system is {round(LenSystem, 3)} people")
    print(f"The expected length of the queue is {round(LenQueue, 3)} people")
    print(f"The expected waiting time of the system is {WaitTimeSystem} hours")
    print(f"The expected waiting time of the queue is {WaitTimeQueue} hours")
    
    option = int(input(f"Are you looking for the probability of 1.exactly 2.At least 3.At most {n} people? (select corresponding numbers)"))
    if option == 1:
        print(f"The probability of {n} people in the system is {prob_of_exact(n, rho, N) * 100}%")
    elif option == 2:
        print(f"The probability of at least {n} people in the system is {prob_of_atleast(n, rho, N) * 100}%")
    elif option == 3:
        print(f"The probability of at most {n} people in the system is {prob_of_atmost(n, rho, N) * 100}%")


def prob_of_exact(num, rho, N=None):
    if N is None:
        return (rho ** num) * (1 - rho)
    else:
        return (rho ** num) * ((1 - rho) / (1 - (rho ** (N + 1))))

def prob_of_atleast(num, rho, N=None):
    if N is None:
        sum_result = sum((rho ** i) * (1 - rho) for i in range(num))
    else:
        sum_result = sum((rho ** i) * (1 - rho) / (1 - (rho ** (N + 1))) for i in range(num))
    return 1 - sum_result

def prob_of_atmost(num, rho, N=None):
    if N is None:
        sum_result = sum((rho ** i) * (1 - rho) for i in range(num + 1))
    else:
        sum_result = sum((rho ** i) * (1 - rho) / (1 - (rho ** (N + 1))) for i in range(num + 1))
    return sum_result

def main():
    queue_type = int(input("Is it an infinite(1) queue or a finite(2) queue? (select corresponding number)"))
    if queue_type == 1:
        arrival_rate = int(input("What is the arrival rate in your system? "))
        service_rate = int(input("What is the service rate in your system? "))
        n = int(input("\nWhat probability of people in the system are you looking for? "))
        infinite_queue(arrival_rate, service_rate, n)
        while True:
            c = input("Do you want to query again? (yes/no)")
            if c.lower() != "yes":
                break
            n = int(input("\nWhat probability of people in the system are you looking for? "))
            infinite_queue(arrival_rate, service_rate, n)
    elif queue_type == 2:
        N = int(input("What is the limit of people that can be in the system? "))
        arrival_rate = int(input("What is the arrival rate in your system? "))
        service_rate = int(input("What is the service rate in your system? "))
        n = int(input("\nWhat probability of people in the system are you looking for? "))
        finite_queue(arrival_rate, service_rate, n, N)
        while True:
            c = input("Do you want to query again? (yes/no)")
            if c.lower() != "yes":
                break
            n = int(input("\nWhat probability of people in the system are you looking for? "))
            finite_queue(arrival_rate, service_rate, n, N)
    else:
        print("Invalid Option!!")

if __name__ == "__main__":
    main()
```
